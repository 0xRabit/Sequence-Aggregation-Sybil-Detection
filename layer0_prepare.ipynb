## 大文件按月拆分

```python
import csv
from datetime import datetime

# 输入和输出文件路径
input_file_path = f'layer0/{NROWS}.csv'
input_file_path = "layer0/2024-05-15-snapshot1_transactions.csv" #1.5h

# 函数用于打开一个新的CSV文件并返回文件对象和写入器
def open_monthly_file(month_key, fieldnames):
    output_file = open(f'layer0/bak/{month_key}.csv', mode='w', newline='', encoding='utf-8')
    writer = csv.DictWriter(output_file, fieldnames=fieldnames)
    writer.writeheader()
    return output_file, writer

# 使用流式读取处理大型CSV文件
with open(input_file_path, mode='r', encoding='utf-8') as csvfile:
    reader = csv.DictReader(csvfile)
    
    # 字典用于存储打开的文件句柄和写入器
    file_handles = {}
    
    for row in reader:
        # 解析时间戳字段并按照给定的格式转换为datetime对象
        timestamp = datetime.strptime(row['SOURCE_TIMESTAMP_UTC'], '%Y-%m-%d %H:%M:%S.%f')
        # 生成月份的键值
        month_key = timestamp.strftime('%Y%m')

        # 如果这个月份的文件还没有打开过，打开它
        if month_key not in file_handles:
            file_handles[month_key] = open_monthly_file(month_key, reader.fieldnames)

        # 获取文件写入器并写入行
        _, writer = file_handles[month_key]
        writer.writerow(row)

# 关闭所有打开的文件句柄
for file, _ in file_handles.values():
    file.close()

print("CSV files have been created for each month.")
```

## 处理某一天的切片

```python
##定义函数process_one_day(df,day),输入YYYYMMDD格式，截取SOURCE_TIMESTAMP_UTC字段5AM-17PM的数据
def process_one_day(csv_file, day):
  """
  Extracts data from the specified day and time range.

  Args:
      df: Pandas DataFrame containing the data.
      day: String in YYYYMMDD format representing the target day.

  Returns:
      Pandas DataFrame containing the filtered data.
  """
  df = pd.read_csv(csv_file,index_col=0)

  # Convert the day string to datetime format
  day_datetime = pd.to_datetime(day)

  # Convert the SOURCE_TIMESTAMP_UTC column to datetime format
  df['SOURCE_TIMESTAMP_UTC'] = pd.to_datetime(df['SOURCE_TIMESTAMP_UTC'])

  # Create a boolean mask to filter data between 5am and 5pm
  mask = (df['SOURCE_TIMESTAMP_UTC'].dt.hour >= 5) & (df['SOURCE_TIMESTAMP_UTC'].dt.hour < 17) & (df['SOURCE_TIMESTAMP_UTC'].dt.date == day_datetime.date())

  # Filter the DataFrame based on the mask
  filtered_df = df[mask]

  return filtered_df

process_one_day('layer0/bak/202401.csv','20240101')  
```
